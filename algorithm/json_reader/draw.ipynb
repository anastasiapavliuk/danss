```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
with open('f194e830714.json', 'r') as f:
    data = json.load(f)
```

```python
data['x']
data['z_x']
data['k_x']
data['c_x']
```

```python
x = data['x']
z_x = data['z_x']

zs = np.linspace(min(z_x),max(z_x), 100)
xs = data['k_x']*zs + data['c_x']

fig, axs = plt.subplots(figsize=(12,12))
plt.hist2d(x=x, y=z_x, bins=[25,100], range=[[0, 100], [0, 104]], cmin=1e-24)
plt.plot(xs, zs, c='cyan', lw=3, ls='--')
plt.xlabel("X")
plt.ylabel("Z")
plt.xlim([0,100])
plt.ylim([0,104])
fig.savefig("histXZ.png")

y = data['y']
z_y = data['z_y']

zs = np.linspace(min(z_y),max(z_y), 100)
ys = data['k_y']*zs + data['c_y']

fig, axs = plt.subplots(figsize=(12,12))
plt.hist2d(x=y, y=z_y, bins=[25,100], range=[[0, 100], [0, 104]], cmin=1e-24)
plt.plot(ys, zs, c='cyan', lw=3, ls='--')
plt.xlabel("Y")
plt.ylabel("Z")
plt.xlim([0,100])
plt.ylim([0,104])
fig.savefig("histYZ.png")
```
