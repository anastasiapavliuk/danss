```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
with open('mce7916.json', 'r') as f:
    data = json.load(f)
```

```python
data['x']
data['z_x']
data['k_x']
data['c_x']
data['y']
data['z_y']
data['k_y']
data['c_y'] 
```

```python
with open('mc0e7916.json', 'r') as f:
    data0 = json.load(f)
```

```python
data0['x']
data0['z_x']
data0['y']
data0['z_y']
```

График по восстановленным стрипам

```python
x = data['x']
z_x = data['z_x']

zs = np.linspace(min(z_x),max(z_x), 100)
xs = data['k_x']*zs + data['c_x']

y = data['y']
z_y = data['z_y']

zs = np.linspace(min(z_y),max(z_y), 100)
ys = data['k_y']*zs + data['c_y']

fig, axs = plt.subplots(figsize=(16,8), nrows=1 , ncols=2)
axs[0].hist2d(x=x, y=z_x, bins=[25,100], range=[[0, 100], [0, 104]], cmin=1e-24)
axs[0].plot(xs, zs, c='cyan', lw=3, ls='--')
axs[0].set_xlabel("X")
axs[0].set_ylabel("Z")
# axs[0].xlim([0,100])
# axs[0].ylim([0,104])

axs[1].hist2d(x=y, y=z_y, bins=[25,100], range=[[0, 100], [0, 104]], cmin=1e-24)
axs[1].plot(ys, zs, c='cyan', lw=3, ls='--')
axs[1].set_xlabel("Y")
axs[1].set_ylabel("Z")
# axs[1].xlim([0,100])
# axs[1].ylim([0,104])
fig.savefig("histXYZ.png")
```
