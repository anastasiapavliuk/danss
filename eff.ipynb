```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
DATA = pd.read_table('stat_pos_2023.txt',sep='\s+', engine='python')
```

```python
colors = plt.cm.viridis(DATA["Eff"]/np.max(DATA["Eff"]))
colors = sns.color_palette(colors, as_cmap=True)

fig, axs = plt.subplots(figsize=(len(DATA["Eff"]), 40), constrained_layout=True)
axs = sns.barplot(x="Begin", y="Eff", data=DATA, palette=colors, dodge=False)
plt.tick_params(axis='both', which='major', labelsize=20)

plt.grid(True)
fig.savefig('effect.pdf')
```
